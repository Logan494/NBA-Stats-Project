```python
import pandas as pd

# Merge based on exact match on 'age', 'experience', and 'gp'
merged_df = pd.merge(df, df2[['age', 'experience', 'gp', 'mp']],
                     on=['age', 'experience', 'gp'],
                     how='left',  # Left join to keep all rows from df
                     suffixes=('_df', '_df2'))  # Add suffixes to identify columns from each dataframe

# Now you have a merged DataFrame with 'mp' column from df2 merged into df based on exact matches for 'age', 'experience', and 'gp'
```

```python
import pandas as pd

# Drop duplicates from df2 based on 'age', 'experience', 'gp'
df2_no_duplicates = df2.drop_duplicates(subset=['age', 'experience', 'gp'])

# Merge based on exact match on 'age', 'experience', and 'gp'
merged_df = pd.merge(df, df2_no_duplicates[['age', 'experience', 'gp', 'mp']],
                     on=['age', 'experience', 'gp'],
                     how='left')  # Left join to keep all rows from df

# Now you have a merged DataFrame with 'mp' column from df2 merged into df based on exact matches for 'age', 'experience', and 'gp'
```

```python
# Find the number of null values in each column of merged_df
null_values = merged_df.isnull().sum()

# Print the number of null values for each column
print(null_values)
```
